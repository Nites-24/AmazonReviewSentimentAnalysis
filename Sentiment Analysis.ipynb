```python
from transformers import pipeline

# Load sentiment analysis pipeline on CPU
sentiment_pipeline = pipeline("sentiment-analysis", device=-1)  # Force CPU usage
print("Pipeline initialized on CPU.")
```

```python
from transformers import pipeline

# Specify the sentiment analysis model explicitly
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert/distilbert-base-uncased-finetuned-sst-2-english",
    device=-1  # Use CPU
)
print("Pipeline initialized with explicit model on CPU.")
```

```python
# Randomly sample 10,000 rows from the training dataset
sample_train_df = train_df.sample(10000, random_state=42)  # Ensures reproducibility
print(f"Selected {len(sample_train_df)} rows for sentiment analysis.")
```

```python
print("Available columns in the Test DataFrame:")
print(test_df.columns)

print("\nAvailable columns in the Train DataFrame:")
print(train_df.columns)
```

```python
print(sample_train_df.columns)
```

```python
# Add an empty column for predicted sentiment
sample_train_df['predicted_sentiment'] = ""
print("Column 'predicted_sentiment' added.")
print(sample_train_df.columns)  # Verify the column exists
```

```python
batch_size = 100  # Use smaller batches for faster processing

# Process sentiment analysis in batches
for i in range(0, len(sample_train_df), batch_size):
    # Get a batch of reviews
    batch = sample_train_df['content'].iloc[i:i + batch_size].tolist()

    # Run sentiment analysis
    results = sentiment_pipeline(batch)

    # Extract sentiment labels
    sentiments = [result['label'] for result in results]

    # Assign predictions to the 'predicted_sentiment' column
    sample_train_df.iloc[i:i + batch_size, sample_train_df.columns.get_loc('predicted_sentiment')] = sentiments

    # Print progress every 1000 rows
    if i % 1000 == 0:
        print(f"Processed {i} rows...")

print("Sentiment analysis completed!")
```

```python
# Add a column to the sampled dataset
sample_train_df['predicted_sentiment'] = ""

batch_size = 100  # Adjust based on performance
for i in range(0, len(sample_train_df), batch_size):
    # Get a batch of reviews from the sampled dataset
    batch = sample_train_df['content'].iloc[i:i + batch_size].tolist()

    # Run sentiment analysis
    results = sentiment_pipeline(batch)

    # Extract sentiment labels
    sentiments = [result['label'] for result in results]

    # Assign predictions to the 'predicted_sentiment' column in the sampled dataset
    sample_train_df.iloc[i:i + batch_size, sample_train_df.columns.get_loc('predicted_sentiment')] = sentiments

    # Print progress every 1000 rows
    if i % 1000 == 0:
        print(f"Processed {i} rows...")

print("Sentiment analysis completed for the sampled dataset!")
```

```python
# Save the processed dataset
sample_train_df.to_csv("processed_10000_rows.csv", index=False)
print("Processed dataset saved as 'processed_10000_rows.csv'.")
```

```python
import matplotlib.pyplot as plt

# Plot sentiment distribution
sample_train_df['predicted_sentiment'].value_counts().plot(kind='bar', color=['green', 'red', 'blue'])
plt.title("Predicted Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
# Map numeric labels to sentiment text (e.g., 0 -> "NEGATIVE", 1 -> "POSITIVE")
label_to_sentiment = {0: "NEGATIVE", 1: "POSITIVE"}
sample_train_df['label_mapped'] = sample_train_df['label'].map(label_to_sentiment)

# Compare predictions to ground truth and calculate accuracy
accuracy = (sample_train_df['predicted_sentiment'] == sample_train_df['label_mapped']).mean()
print(f"Model Accuracy: {accuracy:.2f}")
```
